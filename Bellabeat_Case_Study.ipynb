Merging data to consolidate analysis

```python
## Rename ActivityDate to Date in the DailyActivity tabel
DailyActivity.rename(columns={'ActivityDate': 'Date'}, inplace= True)
```

```python
## Merging the two tables
Merged_activities = pd.merge(DailyActivity, SleepDay, on=['Id', 'Date'], how='outer')
```

### **VISUALISATION**

```python
## Importing the necessaries libraries for visuals
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.scatter(Merged_activities['Calories'], Merged_activities['TotalSteps'])
plt.xlabel('Calories')
plt.xlabel('TotalSteps')
plt.title('Calories vs TotalSteps')
plt.show()
```

- The more active the customers are, the more calories they burn. However, it is important to note that some customers show reluctance or resistance to change

```python
plt.scatter(Merged_activities['TotalMinutesAsleep'], Merged_activities['TotalTimeInBed'])
plt.xlabel('Total Minutes Asleep')
plt.xlabel('Total Time in Bed')
plt.title('Total Minutes Asleep vs Total time in Bed')
plt.show()
```

- Base on the data, there is a positive correlation between total minutes asleep and total time spend on bed

```python
plt.hist(Merged_activities['TotalTimeInBed'], bins= 3)
plt.xlabel('Total Time in Bed')
plt.ylabel('Frequency of Occurence')
plt.title('Total Time in Bed vs Frequency of Occurence')
plt.show()
```

- Some Customers spend long hours on bed

```python
labels = ['VeryActiveMinutes', 'FairlyActiveMinutes', 'LightlyActiveMinutes', 'SedentaryMinutes']
sizes = Merged_activities[['VeryActiveMinutes', 'FairlyActiveMinutes', 'LightlyActiveMinutes', 'SedentaryMinutes']].sum()
colors = ["lightcoral", "yellowgreen", "lightskyblue", "darkorange"]

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%')
plt.title('Distribution of Active Minutes')
plt.show()

```

- The Data shows a high percentage of inactive state
- Customers are lightly active.

```python
Merged_activities['Date'] = pd.to_datetime(Merged_activities['Date'])
Merged_activities['weekday'] = Merged_activities['Date'].dt.weekday
plt.bar(Merged_activities['weekday'], Merged_activities['VeryActiveMinutes'])
plt.xlabel('Weekday')
plt.ylabel('Very Active Minutes')
plt.title('Very Active by Weekday')
plt.show()
```

- Customers are Highly active during weekends

```python
plt.hist(WEightLoginfo['WeightKg'], bins = 3)
plt.xlabel('Weight')
plt.ylabel('Frequency')
plt.title('Weight Distribution')
plt.show()
```

- It seems like there are some outliers on both ends of weight distribution, but the central tendency is arround 72kg
